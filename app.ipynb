```python
import pandas as pd
import os
```

```python
df = pd.read_csv("hello.csv")
df['TOPIC'].unique()
```

```python
df
```

```python
df = df[df.TOPIC.isin(['bass', 'battery', 'call','case'])]
```

```python
df
```

```python
def fun(x):
    if (x == "bass" or x == "battery"):
        return "hardware"
    elif (x == "call"):
        return "anything"
    else:
        return "software"
```

```python
df["category"] = df.TOPIC.apply(lambda x: fun(x))
```

```python
def ratingfun(x):
    if (x > 0.21):
        return 5
    else:
        return 1
```

```python
df["rating"] = df["SENTIMENT SCORE"].apply(lambda x: ratingfun(x))
```
